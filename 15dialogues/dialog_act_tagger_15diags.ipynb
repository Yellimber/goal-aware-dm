```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
import csv

lne = []
res = []
s = ''
with open("test_set.txt", 'r', encoding = "utf-8") as f:
    for line in f:
        lne = []
        s = ''
        lne.append(line.split('|')[2])
        lne.append(line.split('|')[0])
        doc = nlp(line.split('|')[1])
        for token in doc:
            s += str(token) + '/' + str(token.pos_) + ' '
        if (len(s) == 0):
            s = './.'
        lne.append(s)
        lne.append(line.split('|')[1])
        res.append(lne)
print(res[:5])
```

```python
with open("csv_test/test_set.csv", 'w', encoding = "utf-8", newline = '') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = ["act_tag", "speaker", "pos", "text"])
    writer.writeheader()
    for line in res:
        writer.writerow({"act_tag": line[0], "speaker": line[1], "pos": line[2], "text": line[2]})
```

```python
lne = []
res = []
s = ''
with open("train_set.txt", 'r', encoding = "utf-8") as f:
    for line in f:
        lne = []
        s = ''
        lne.append(line.split('|')[2])
        lne.append(line.split('|')[0])
        doc = nlp(line.split('|')[1])
        for token in doc:
            s += str(token) + '/' + str(token.pos_) + ' '
        if (len(s) == 0):
            s = './.'
        lne.append(s)
        lne.append(line.split('|')[1])
        res.append(lne)
print(res[:5])
```

```python
with open("csv_train/train_set.csv", 'w', encoding = "utf-8", newline = '') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = ["act_tag", "speaker", "pos", "text"])
    writer.writeheader()
    for line in res:
        writer.writerow({"act_tag": line[0], "speaker": line[1], "pos": line[2], "text": line[2]})
```
